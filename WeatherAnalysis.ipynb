```python
# Dependencies
from config import api_key
from citipy import citipy
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import requests as req
```

```python
lats = np.random.uniform(low=-90.000, high=90.000, size=500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=500)
base_weather_url = "http://api.openweathermap.org/data/2.5/weather?"
# Build query URL
query_url = base_weather_url + "appid=" + api_key + "&q="
```

```python
lats
```

```python
cities = []
```

```python
for lat, lang in zip(lats, lngs):
    print(lat,lang)
    cities.append(citipy.nearest_city(lat,lang).city_name)
```

```python
cities
```

```python
print(cities)
```

```python
city_weather = []
for city in cities:
    print("City Name::",city)
    url = query_url+city
    print(url)
    response = req.get(query_url+city)
    print(response.json())
    city_weather.append(response.json())
```

```python
temperature=[]
pressure=[]
humidity=[]
cloudiness=[]
wind_speed = []
latitude=[]
for weather in city_weather:
    temperature.append(weather['main']['temp'])
    pressure.append(weather['main']['pressure'])
    humidity.append(weather['main']['humidity'])
    cloudiness.append(weather['clouds']['all'])
    wind_speed.append(weather['wind']['speed'])
    latitude.append(weather['coord']['lat'])
print("Temperature::",temperature)
print("Pressure::",pressure)
print("Humidity::",humidity)
print("Cloudiness::",cloudiness)
print("WindSpeed::",wind_speed)
print("Latitude::",latitude)
```

```python
weather_data = pd.DataFrame(list(zip(latitude,temperature, pressure, humidity,cloudiness,wind_speed)),
              columns=['Latitude','Temperature','Pressure', 'Humidity','Cloudiness','Wind Speed'])
```

```python
weather_data.to_csv('WeatherData.csv')
```

```python
#Scatter plot 1 
#Temperature (F) vs. Latitude
plt.scatter(temperature, latitude,edgecolors='black',marker="o", alpha=0.8)
plt.title("Temperature (F) vs. Latitude") 
plt.xlabel("Temperature") 
plt.ylabel("Latitude") 
plt.ylim([-90, 90]) 
plt.grid(True) 
plt.savefig("TempVsLat.png") 
plt.show()




#Humidity (%) vs. Latitude
plt.scatter(humidity, latitude,edgecolors='black',marker="o", alpha=1.0)
plt.title("Humidity (%) vs. Latitude") 
plt.xlabel("Humidity") 
plt.ylabel("Latitude") 
plt.ylim([-90, 90]) 
plt.grid(True) 
plt.savefig("HumidVsLat.png") 
plt.show()
# Cloudiness (%) vs. Latitude
plt.scatter(cloudiness, latitude,edgecolors='black',marker="o", alpha=1.0)
plt.title("Cloudiness (%) vs. Latitude") 
plt.xlabel("Cloudiness") 
plt.ylabel("Latitude") 
plt.ylim([-90, 90]) 
plt.grid(True) 
plt.savefig("CloudVsLat.png") 
plt.show()
#Wind Speed (mph) vs. Latitude
plt.scatter(wind_speed, latitude,edgecolors='black',marker="o", alpha=1.0)
plt.title("Wind Speed (mph) vs. Latitude") 
plt.xlabel("Wind Speed") 
plt.ylabel("Latitude") 
plt.ylim([-90, 90]) 
plt.grid(True) 
plt.savefig("WindSpVsLat.png") 
plt.show()
```
